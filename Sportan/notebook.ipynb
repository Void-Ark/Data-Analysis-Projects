```python
import pandas as pd 
import json 
```

```python
with open('./dataset_json/t20_wc_match_results.json', 'r') as file : 
    data = json.load(file)
df_result = pd.DataFrame(data[0]['matchSummary'])
df_result.rename({'scorecard' : 'match_id'}, axis=1, inplace=True)
df_result.head()
```

```python
with open('./dataset_json/t20_wc_batting_summary.json', 'r') as file : 
    data = json.load(file) 

    all_record = [] 
    
    for rec in data : 
        all_record.extend(rec['battingSummary']) 
df_batting = pd.DataFrame(all_record) 
df_batting.head()
```

```python
# making a new column for out and not out using dismissal 
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x : 'out' if x else 'not_out') 
df_batting.head(11)
```

```python
# dropping the not needed column dismissal 
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting.head()
```

```python
# removing special characters from batsman name
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('\xa0', ''))
df_batting.head()
```

```python
# adding match_id in df_batsman from df_result
match_id_dict = {}
for index, row in df_result.iterrows() : 
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1'] 
    
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']
    
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
df_batting.head()
```

```python
# saving the csv file of df_batting dataset
df_batting.to_csv('./cleaned_dataset_csv/fact_bating_summary.csv', index=False)
```
